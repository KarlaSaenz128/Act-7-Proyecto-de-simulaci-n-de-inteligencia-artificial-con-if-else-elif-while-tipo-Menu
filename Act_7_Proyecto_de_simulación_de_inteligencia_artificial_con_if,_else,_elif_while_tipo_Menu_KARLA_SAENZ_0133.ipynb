```python
print("Programa realizado por Nicole Robles NL = 50")
print("Bienvenido al Cerebro Lógico de IA para Influensers")

opcion_menu = "0"
while opcion_menu != "3":
    print("\n--- MENÚ DE EVALUACIÓN DE IA PARA INFLUENSERS ---")
    print("1. Evaluar Perfil de Productor de Contenido")
    print("2. Evaluar Perfil de Streamer en Vivo")
    print("3. Salir del Sistema")

    opcion_menu = input("\nSeleccione una opción (1-3): ")

    if opcion_menu == "1":
        print("\n[MODO IA: EVALUACIÓN DE PRODUCTOR]")
        nombre = input("Nombre del influenser: ")
        while True:
            try:
                experiencia = int(input("Años de trayectoria en redes: "))
                break
            except ValueError:
                print("Entrada inválida. Por favor, ingrese un número para los años de trayectoria.")
        engagement = float(input("Tasa de engagement (0.0 a 10.0): "))
        nicho = input("¿Nicho 'tecnologia' o 'belleza'?: ").lower()

        # Ruta 1: Review Tecnológico Avanzado
        if nicho == "tecnologia" and engagement >= 7.5:
            print(f"IA: {nombre}, Recomendación: Video-Review a profundidad de Gadgets Premium (Alto Impacto).")

        # Ruta 2: Contenido Corto/Virales de Tech
        elif nicho == "tecnologia" and engagement < 7.5:
            print(f"IA: {nombre}, Recomendación: Shorts/Reels rápidos con tendencias tecnológicas (Engagement Moderado).")

        # Ruta 3: Tutoriales Premium o Campañas de Belleza
        elif nicho == "belleza" or experiencia > 10:
            print(f"IA: {nombre}, Recomendación: Masterclass de Skincare avanzado o Campaña con marcas internacionales.")

        # Ruta 4: Blogs de Estilo de Vida
        else:
            print(f"IA: {nombre}, Recomendación: Vlogs diarios de estilo de vida y dinámicas básicas con fans.")

    elif opcion_menu == "2":
        print("\n[MODO IA: EVALUACIÓN DE STREAMER]")
        nombre = input("Nombre del influenser: ")
        plataforma = input("Plataforma principal ('twitch', 'tiktok' u 'youtube'): ").lower()
        while True:
            try:
                frecuencia_semanal = int(input("¿Días que transmites por semana? (1 a 7): "))
                break
            except ValueError:
                print("Entrada inválida. Por favor, ingrese un número para los días de transmisión.")
        engagement = float(input("Tasa de engagement (0.0 a 10.0): "))

        # Ruta 5: Twitch Especializado
        if plataforma == "twitch" and frecuencia_semanal >= 4:
            print(f"IA: {nombre}, Recomendación: Evento especial comunitario o Maratón de transmisión (Subathon).")

        # Ruta 6: Tiktok In Live
        elif plataforma == "tiktok" and (frecuencia_semanal >= 3 or engagement > 8.0):
            print(f"IA: {nombre}, Recomendación: Batallas en directo o Dinámicas interactivas virales (Potencia).")

        # Ruta 7: YouTube Lives
        elif plataforma == "youtube" and frecuencia_semanal >= 2:
            print(f"IA: {nombre}, Recomendación: Podcast interactivo en vivo o Análisis en tiempo real (Comercial).")

        # Ruta 8: Optimización de Configuración básica
        else:
            print(f"IA: {nombre}, Recomendación: Ajustar configuraciones de audio/video y planificar horarios.")

    elif opcion_menu == "3":
        print("\nSaliendo del sistema...")

    # Ruta 9: Error de sistema
    else:
        print("\nIA: Error. Opción de menú no válida.")

print("\n--- Sistema Finalizado ---")
```
